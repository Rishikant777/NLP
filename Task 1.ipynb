# NLP Preprocessing Engine

## Task 1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?  
In NLP, both can be treated as different words if we do not convert text to lowercase. Because of that, the model may think they are two separate tokens. Lowercasing helps make the text consistent.

2. What happens if stopwords are not removed?  
If stopwords are not removed, the text may contain many common words like “is”, “the”, “and”, which do not add much meaning. This can create noise and reduce the importance of useful words.

3. Provide two real-world scenarios where removing stopwords can be harmful.  
First, in sentiment analysis, words like “not” and “no” are very important. For example, “not good” becomes “good” if “not” is removed, which changes the meaning.  
Second, in chatbot or question-answer systems, removing stopwords can sometimes break the meaning of the sentence and make the response less accurate.

4. Explain the difference between stemming and lemmatization.  
Stemming cuts a word into a root form using simple rules, so sometimes the output is not a proper word. For example, “studies” may become “studi”.  
Lemmatization converts a word into its correct base form using language rules. For example, “studies” becomes “study”.

```python
import re
from collections import Counter

stopwords = {
    "a","an","the","is","are","am","was","were","be","been","being",
    "and","or","but","if","then","than","so","of","at","by","for",
    "to","from","with","on","in","into","out","as","it","this","that",
    "these","those","he","she","they","we","you","your","our","their",
    "his","her","its","i","me","my","mine","us","them"
}

keep_words = {"no", "not"}

def reduce_repeated_chars(word):
    return re.sub(r'(.)\1{2,}', r'\1', word)

def preprocess_text(text):
    if not text:
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', ' ', text)
    text = re.sub(r'\S+@\S+', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    tokens = []

    for word in words:
        word = reduce_repeated_chars(word)

        if len(word) <= 2 and word not in keep_words:
            continue

        if word in stopwords and word not in keep_words:
            continue

        tokens.append(word)

    return tokens, " ".join(tokens)
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for i, s in enumerate(test_sentences, 1):
    tokens, clean = preprocess_text(s)
    all_tokens.extend(tokens)

    print("Sentence", i)
    print("Original :", s)
    print("Tokens   :", tokens)
    print("Cleaned  :", clean)
    print("-"*50)
```

```python
for i, s in enumerate(test_sentences, 1):
    tokens, _ = preprocess_text(s)

    total = len(tokens)
    unique = len(set(tokens))
    avg = sum(len(t) for t in tokens)/total if total else 0

    print("Sentence", i)
    print("Total Tokens :", total)
    print("Unique Tokens:", unique)
    print("Avg Length   :", round(avg,2))
    print("-"*50)
```

## Analysis

The sentence "Win $$$ now!!! Limited offer!!!" contains the most noise because it includes symbols, repeated punctuation, and less meaningful words.

The sentence "I absolutely looooved this product 😍" retains meaningful tokens like "absolutely", "loved", and "product" after cleaning.

```python
from collections import Counter

freq = Counter(all_tokens)

print("Top 10 words:")
for w, c in freq.most_common(10):
    print(w, ":", c)

print("\nLeast 5 words:")
for w, c in sorted(freq.items(), key=lambda x:x[1])[:5]:
    print(w, ":", c)
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for t in text_list:
        tokens, clean = preprocess_text(t)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
print(full_pipeline(test_sentences))
```
